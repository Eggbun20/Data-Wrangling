# TẠO FEATURES GỒM CÁC CỘT CẦN THIẾT

1. UNITID : UNIT ID của tổ chức
2. INSTNM : Tên trường học
3. CITY : Thành phố
4. STABBR : Bang
5. LOCALE : Vị trí phân bổ
6. PCIP11 : Phần trăm các bằng cấp được trao của ngành Computer And Information Sciences And Support Services.
7. CIP11CERT1 : Chứng chỉ dưới một năm học về Computer And Information Sciences And Support Services.
8. CIP11CERT2 : Chứng chỉ của ít nhất một nhưng dưới hai năm học về Computer And Information Sciences And Support Services.
9. CIP11ASSOC : Bằng Cao Đẳng về Computer And Information Sciences And Support Services.
10. CIP11CERT4 : Giải thưởng của hơn hai nhưng dưới bốn năm học về Computer And Information Sciences And Support Services.
11. CIP11BACHL : Bằng Cử Nhân về Computer And Information Sciences And Support Services.

```python
features = ['UNITID','INSTNM','CITY','STABBR','LOCALE','PCIP11','CIP11CERT1','CIP11CERT2',
'CIP11ASSOC','CIP11CERT4','CIP11BACHL']
```

# XOÁ NA/NULL VÀ CHỌN DỮ LIỆU
- Xoá các dữ liệu Null trong cột PCIP11 và lấy dữ liệu # 0.0
- Cột LOCALE chỉ lấy 11 , 12 , 13 vì đó là các thành phố (theo đề yêu cầu)

```python
pd.set_option('display.max_rows', 30)
university_features = university[features]
university_features.dropna(subset = ['PCIP11']).query("LOCALE >= 11 & LOCALE <= 13").query("PCIP11 > 0")
```

# Tạo cột Total of crime cho Crime

```python
crime['Total of crime'] = crime.sum(axis=1)
```

```python
crime.groupby('City').mean()
```

# Thành phố 50th percentile vể tổng số vụ phạm tội 

```python
crime['Total of crime'].quantile(0.5)
```

```python
citys_crime = crime.loc[crime['Total of crime'] <= 5871].groupby('City').mean().sort_values('Total of crime', ascending=True)
citys_crime.head(10)
```
